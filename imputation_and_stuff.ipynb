```python
# Let's manipulate this data so that "Yes" "No" can be binary, but first, lets copy the data

# Copy the data to manipulate
dropout_df_copy = dropout_df.copy()
```

```python
# Now that we have a copy, lets go ahead and clean this up

# Convert columns with "yes/no" to "1,2"
dropout_df_copy.update(dropout_df_copy.Withdrew.replace('Yes', 1))
dropout_df_copy.update(dropout_df_copy.Withdrew.replace('No', 2))

# Check Data now
dropout_df_copy.update(dropout_df_copy.Withdrew)
```

```python
# Good, now let's build our model

# What we're trying to predict
drop_out = dropout_df_copy.Withdrew
drop_out_predictors = dropout_df_copy.drop(['Withdrew'], axis=1)

# The factors that impact our predictor (numeric predictors)
predictors_for_drop_out = drop_out_predictors.select_dtypes(exclude=['object'])
```

```python
# To prevent us from building our model and testing it with one dataset, lets split ours into various samples

# Import sklearn data split
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Split our Data!
model_predictors_for_drop_out, test_predictors_for_drop_out, model_drop_out, test_drop_out = train_test_split(predictors_for_drop_out, drop_out, train_size=0.7, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Now, let's do this: Let's score our data based on the different ways to handling missing values

# Scored Data from dropping columns with missing values

columns_with_missing_values = [col for col in model_predictors_for_drop_out.columns
                              if model_predictors_for_drop_out[col].isnull().any()]

reduced_model_predictors_for_drop_out = model_predictors_for_drop_out.drop(columns_with_missing_values, axis=1)
reduced_test_predictors_for_drop_out = test_predictors_for_drop_out.drop(columns_with_missing_values,axis=1)

# Now lets define a function to score our dataset
def score_dataset(X_train, X_test, y_train, y_test):
    model = RandomForestRegressor()
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return mean_absolute_error(y_test, preds)
```

```python
test_drop_out.update(test_drop_out.fillna(1))
```

```python
print('Mean Absolute Error from dropping columns with Missing Values:')
print(score_dataset(reduced_model_predictors_for_drop_out, reduced_test_predictors_for_drop_out, model_drop_out, test_drop_out.dropna()))
```

```python
# Now that we've gotten our MAE with dropping columns. Let's give Imputation a try!

# Import Imputer from sklearn
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer()

# Impute the data!
imputed_model_predictors_for_drop_out = my_imputer.fit_transform(model_predictors_for_drop_out)
imputed_test_predictors_for_drop_out = my_imputer.transform(test_predictors_for_drop_out)
```

```python
print('Mean Absolute Error from Imputation:')
print(score_dataset(imputed_model_predictors_for_drop_out, imputed_test_predictors_for_drop_out, model_drop_out, test_drop_out.dropna()))
```

```python
# Now lets impute our data but lets tag what was imputed
plus_model_predictors_for_drop_out = model_predictors_for_drop_out.copy()
plus_test_predictors_for_drop_out = test_predictors_for_drop_out.copy()

# Get missing columns
columns_with_missing_values = (col for col in model_predictors_for_drop_out.columns
                              if model_predictors_for_drop_out[col].isnull().any())

# Add new Column and flag missing values
for col in columns_with_missing_values:
    plus_model_predictors_for_drop_out[col + "_imputed"] = plus_model_predictors_for_drop_out[col].isnull()
    plus_test_predictors_for_drop_out[col + "_imputed"] = plus_test_predictors_for_drop_out[col].isnull()
    
# IMPUTE
my_imputer = SimpleImputer()

plus_imputed_model_predictors_for_drop_out = my_imputer.fit_transform(plus_model_predictors_for_drop_out)
plus_imputed_test_predictors_for_drop_out = my_imputer.transform(plus_test_predictors_for_drop_out)
```

```python
# Now print these variables out to see what happens!

print('Mean Absolute Error from Imputation Flagging Imputed:')
print(score_dataset(plus_imputed_model_predictors_for_drop_out, plus_imputed_test_predictors_for_drop_out, model_drop_out, test_drop_out.dropna()))
```
